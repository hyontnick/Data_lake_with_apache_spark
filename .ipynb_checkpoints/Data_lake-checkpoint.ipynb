# 📌 Stratégie de Nettoyage Minimal

### 📌 a) Nettoyage des Données Relationnelles (SQL)

```python
from pyspark.sql.functions import col, to_date, lower

# Suppression des NULLs et normalisation des emails
clients_clean_df = clients_df.dropna().withColumn("email", lower(col("email")))

# Nettoyage des transactions : suppression des NULLs et formatage des dates
transactions_clean_df = transactions_df.dropna().withColumn("date_achat", to_date(col("date_achat"), "yyyy-MM-dd"))

# Sauvegarde des données nettoyées
clients_clean_df.write.mode("overwrite").parquet("data_lake/cleansed/clients/")
transactions_clean_df.write.mode("overwrite").parquet("data_lake/cleansed/transactions/")

print("✅ Données SQL nettoyées et stockées dans Cleansed Zone !")
```

### 📌 b) Nettoyage des Données Semi-Structurées (CSV & JSON)


```python
# Suppression des NULLs et correction des prix négatifs
produits_clean_df = produits_df.dropna().filter(col("prix") > 0)

# Suppression des NULLs et formatage des dates pour commandes
commandes_clean_df = commandes_df.dropna().withColumn("date", to_date(col("date"), "yyyy-MM-dd"))

# Sauvegarde
produits_clean_df.write.mode("overwrite").parquet("data_lake/cleansed/produits/")
commandes_clean_df.write.mode("overwrite").parquet("data_lake/cleansed/commandes/")
print("✅ Données CSV et JSON nettoyées et stockées !")
```

### 📌 c) Nettoyage des Fichiers Texte (Avis Clients & Logs Serveur)

```python
# Nettoyage Avis Clients (Suppression des lignes vides et doublons)
avis_clean_rdd = avis_rdd.filter(lambda x: x.strip() != "").distinct()
avis_clean_rdd.saveAsTextFile("data_lake/cleansed/avis_clients.txt")

# Nettoyage Logs (Suppression des lignes vides et doublons)
logs_clean_rdd = logs_rdd.filter(lambda x: x.strip() != "").distinct()
logs_clean_rdd.saveAsTextFile("data_lake/cleansed/logs/serveur_clean.log")

print("✅ Fichiers texte et logs nettoyés et stockés !")
```

### 📌 d) Vérification des Images

```python
import os
from PIL import Image

# 📂 Créer un dossier pour les images corrompues
os.makedirs("data_lake/errors", exist_ok=True)

image_folder = "data_lake/raw/images/"
clean_folder = "data_lake/cleansed/images/"

os.makedirs(clean_folder, exist_ok=True)

for filename in os.listdir(image_folder):
    if filename.endswith(".png") or filename.endswith(".jpg"):
        try:
            img = Image.open(os.path.join(image_folder, filename))
            img.verify()  # Vérification
            img.close()
            os.rename(os.path.join(image_folder, filename), os.path.join(clean_folder, filename))
        except Exception as e:
            os.rename(os.path.join(image_folder, filename), os.path.join("data_lake/errors", filename))

print("✅ Vérification des images terminée !")
```

# 📊  Analyse et Visualisation

## 📌 Charger les Données Nettoyées dans Spark

```python
# 🔥 Initialisation de Spark
spark = SparkSession.builder.appName("Dashboard").getOrCreate()

# Chargement des données nettoyées
clients_df = spark.read.parquet("data_lake/cleansed/clients/")
transactions_df = spark.read.parquet("data_lake/cleansed/transactions/")
produits_df = spark.read.parquet("data_lake/cleansed/produits/")
commandes_df = spark.read.parquet("data_lake/cleansed/commandes/")
logs_rdd = spark.sparkContext.textFile("data_lake/cleansed/logs/serveur_clean.log")
avis_rdd = spark.sparkContext.textFile("data_lake/cleansed/avis_clients.txt")

print("✅ Toutes les données sont chargées avec succès !")
```

## 📌 2. Effectuer des Analyses

### 💰 a) Chiffre d’Affaires Total

```python
from pyspark.sql.functions import sum

chiffre_affaires = transactions_df.select(sum("montant").alias("chiffre_affaires")).collect()[0]["chiffre_affaires"]
print(f"💰 Chiffre d'affaires total : {chiffre_affaires} €")
```

### 🏆 b) Top 5 des Clients les Plus Rentables

```python
from pyspark.sql.functions import desc

top_clients_df = transactions_df.groupBy("client_id").agg(sum("montant").alias("total_achat")).orderBy(desc("total_achat")).limit(5)
top_clients_df.show()
```

### 📈 c) Produits les Plus Vendus

```python
top_produits_df = commandes_df.groupBy("id").count().orderBy(desc("count")).limit(5)
top_produits_df.show()
```

### 📊 d) Évolution des Ventes par Mois

```python
from pyspark.sql.functions import date_format

ventes_par_mois_df = transactions_df.withColumn("mois", date_format("date_achat", "yyyy-MM")).groupBy("mois").agg(sum("montant").alias("chiffre_affaires")).orderBy("mois")
ventes_par_mois_df.show()
```

## 📌 3. Analyse des Logs Serveur

### 📌 1. Chargement et Extraction des Logs

```python
from pyspark.sql.functions import split, col, count, to_timestamp

# 🔥 Initialisation de Spark
spark = SparkSession.builder.appName("LogAnalysis").getOrCreate()

# 🔹 Chargement des logs sous forme de RDD
logs_rdd = spark.sparkContext.textFile("data_lake/cleansed/logs/serveur_clean.log")

# 🔎 Transformation des logs en DataFrame
logs_df = logs_rdd.map(lambda ligne: ligne.split(" - ")).toDF(["timestamp", "level", "message"])

# 🕒 Conversion des timestamps en format DateTime
logs_df = logs_df.withColumn("timestamp", to_timestamp(col("timestamp"), "yyyy-MM-dd HH:mm:ss.SSSSSS"))

# 📊 Vérification du schéma
logs_df.printSchema()
logs_df.show(5, truncate=False)
```

### 📌 2. Analyse du Nombre de Logs par Niveau

```python
# 🔢 Nombre d'entrées par type de log
logs_count_df = logs_df.groupBy("level").agg(count("*").alias("count")).orderBy(col("count").desc())

# 🖥️ Affichage
logs_count_df.show()
```

### 📌 3. Évolution des Logs au Fil du Temps

```python
from pyspark.sql.functions import date_format

# 📆 Regrouper les logs par minute
logs_par_minute_df = logs_df.withColumn("minute", date_format("timestamp", "yyyy-MM-dd HH:mm")).groupBy("minute").count().orderBy("minute")

# 🔎 Afficher un aperçu
logs_par_minute_df.show(5)
```

### 📌 4. Visualisation avec Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt

# Conversion en Pandas
logs_par_minute_pd = logs_par_minute_df.toPandas()

# 📊 Tracé du graphique
plt.figure(figsize=(12, 5))
plt.plot(logs_par_minute_pd["minute"], logs_par_minute_pd["count"], marker='o', linestyle='-', color="blue")
plt.xticks(rotation=45)
plt.title("Évolution du Nombre de Logs par Minute")
plt.xlabel("Minute")
plt.ylabel("Nombre de Logs")
plt.grid()

# 💾 Sauvegarde de l'image
plt.savefig("evolution_logs.png", dpi=300)  # DPI = Haute qualité

# 📤 Affichage du graphique
plt.show()
```

## 📌 4. Affichage du Tableau de Bord avec Matplotlib & Pandas

```python
# 📊 Chiffre d'Affaires par Mois
ventes_par_mois_pd = ventes_par_mois_df.toPandas()

plt.figure(figsize=(15,5))
plt.plot(ventes_par_mois_pd["mois"], ventes_par_mois_pd["chiffre_affaires"], marker='o', linestyle='-')
plt.xticks(rotation=45)
plt.title("Évolution du Chiffre d'Affaires")
plt.xlabel("Mois")
plt.ylabel("Chiffre d'Affaires (€)")
plt.grid()

# 💾 Sauvegarde de l'image
plt.savefig("chiffre_affaires.png", dpi=300)  # Sauvegarde en haute qualité

# 📤 Affichage
plt.show()
```

## 📌 5. Analyse des Avis Clients

```python
# 🔎 Nombre total d'avis
total_avis = avis_rdd.count()
print(f"📝 Nombre total d'avis clients : {total_avis}")

# 🔎 Top 5 des mots les plus fréquents (excluant mots vides)
from pyspark.sql.functions import lower
import re

# Liste de mots vides courants (on peut l'améliorer)
stopwords = ["le", "la", "les", "un", "une", "de", "du", "des", "et", "est", "à", "pour"]

def nettoyer_avis(avis):
    mots = re.findall(r"\b\w+\b", avis.lower())  # Extraction des mots
    return [mot for mot in mots if mot not in stopwords]

mots_rdd = avis_rdd.flatMap(nettoyer_avis)
mots_count = mots_rdd.map(lambda mot: (mot, 1)).reduceByKey(lambda a, b: a + b).sortBy(lambda x: -x[1])

print("📢 Top 5 des mots les plus fréquents dans les avis :")
for mot, count in mots_count.take(5):
    print(f"{mot} -> {count} occurrences")
```

### 📌 6. 📢 Top 5 des Mots dans les Avis Clients

```python
# 📊 Conversion en DataFrame Pandas
mots_top_pd = pd.DataFrame(mots_count.take(5), columns=["Mot", "Occurrences"])

plt.figure(figsize=(8,5))
plt.bar(mots_top_pd["Mot"], mots_top_pd["Occurrences"], color='orange')
plt.title("Top 5 des Mots Utilisés dans les Avis")
plt.xlabel("Mot")
plt.ylabel("Occurrences")

# 💾 Sauvegarde de l'image
plt.savefig("top_mots_avis.png", dpi=300)  # Enregistrement en haute qualité

# 📤 Affichage
plt.show()
```
